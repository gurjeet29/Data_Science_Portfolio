**Processing the movies till 2016**

```python
#importing modules
import pandas as pd
import numpy as np
```

```python
#reading movie_metadata dataset
data_2016 = pd.read_csv('movie_metadata.csv')
data_2016.head()
```

```python
#recommendation will be based on these features only
data_2016 = data_2016[['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```

**Processing the movies of 2017**

```python
#reading dataset credits
credits = pd.read_csv('credits.csv')
credits.head()
```

```python
#reading dataset movies_metadata
meta = pd.read_csv('movies_metadata.csv')
meta.head()
```

```python
#converting to datetime
meta['release_date'] = pd.to_datetime(meta['release_date'], errors='coerce')
```

```python
#checking number of movies released according to year
meta['year'] = meta['release_date'].dt.year
meta['year'].value_counts().sort_index()
```

```python
#We are getting only 2017 movies as we already have movies up to the year 2016 in preprocessing 1 file(data_2016). 
#We don't have enough data for the movies from 2018, 2019 and 2020. 
#We'll deal with it in the upcoming preprocessing files
new_meta = meta.loc[meta['year'] == 2017,['genres','id','title','year']]
```

```python
new_meta.head()
```

```python
#merging new_meta and credits dataset
new_meta['id'] = new_meta['id'].astype(int)
data_2017 = pd.merge(new_meta, credits, on='id')
```

```python
#converting string to list
import ast
data_2017['genres'] = data_2017['genres'].map(lambda x: ast.literal_eval(x))
data_2017['cast'] = data_2017['cast'].map(lambda x: ast.literal_eval(x))
data_2017['crew'] = data_2017['crew'].map(lambda x: ast.literal_eval(x))
```

```python
def convert(x):
  L = []
  for i in x:
    L.append(i['name'])
    return (" ".join(L))

data_2017['genres'] = data_2017['genres'].map(lambda x: convert(x))
```

```python
#fetching the first 1st actor name
def get_actor1(x):
    casts = []
    for i in x:
        casts.append(i['name'])
    if casts == []:
        return np.NaN
    else:
        return (casts[0])

data_2017['actor_1_name'] = data_2017['cast'].map(lambda x: get_actor1(x))

#fetching the first 2nd actor name
def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i['name'])
    if casts == [] or len(casts)<=1:
        return np.NaN
    else:
        return (casts[1])

data_2017['actor_2_name'] = data_2017['cast'].map(lambda x: get_actor2(x))

#fetching the first 3rd actor name
def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i['name'])
    if casts == [] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])

data_2017['actor_3_name'] = data_2017['cast'].map(lambda x: get_actor3(x))
```

```python
#fetching the director's name
def get_directors(x):
    dt = []
    for i in x:
        if i['job'] == 'Director':
            dt.append(i['name'])
    if dt == []:
        return np.NaN
    else:
        return (" ".join(dt))

data_2017['director_name'] = data_2017['crew'].apply(get_directors)
```

```python
#selecting only necessary features
data_2017 = data_2017[['director_name','actor_1_name','actor_2_name','actor_3_name','genres','title']]
```

```python
data_2017
```

```python
#dropping null values and renaming features
data_2017.isnull().sum()
data_2017.dropna(inplace=True)
data_2017 = data_2017.rename(columns={'title':'movie_title'})
```

```python
data_2017['movie_title'] = data_2017['movie_title'].str.lower()
```

```python
data_2017['comb'] = data_2017['actor_1_name'] + ' ' + data_2017['actor_2_name'] + ' '+ data_2017['actor_3_name'] + ' '+ data_2017['director_name'] +' ' + data_2017['genres']
```

```python
data_2017
```

```python
#merging movies and data datasets
new = data_2016.append(data_2017)
new
```

```python
#dropping the duplicates
new.drop_duplicates(subset ="movie_title", keep = 'last', inplace = True)
new
```

```python
new.to_csv('data_16_17.csv',index=False)
```
