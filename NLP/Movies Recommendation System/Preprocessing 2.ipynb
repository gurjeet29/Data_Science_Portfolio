Extracting features of 2018 movies from Wikipedia

```python
import pandas as pd
import numpy as np
```

```python
def data(link):
  df1 = pd.read_html(link, header=0)[2]
  df2 = pd.read_html(link, header=0)[3]
  df3 = pd.read_html(link, header=0)[4]
  df4 = pd.read_html(link, header=0)[5]

  final_df_2018 = df1.append(df2.append(df3.append(df4,ignore_index=True),ignore_index=True),ignore_index=True)
  return final_df_2018
```

```python
link = "https://en.wikipedia.org/wiki/List_of_American_films_of_2018"
final_df_2018=data(link)
final_df_2018
```

```python
from tmdbv3api import TMDb, Movie
import json
import requests

tmdb = TMDb()
tmdb_movie = Movie()
tmdb.api_key = '22eb0026986397f50a2bbd92a1c791b9'

def get_genre(x):
    genres = []
    result = tmdb_movie.search(x)
    movie_id = result[0].id
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
    data_json = response.json()
    if data_json['genres']:
        for i in range(0,len(data_json['genres'])):
            genres.append(data_json['genres'][i]['name'])
        return (" ".join(genres))
    else:
        np.NaN
```

```python
final_df_2018['genres'] = final_df_2018['Title'].map(lambda x: get_genre(str(x)))
final_df_2018
```

```python
df_2018 = final_df_2018[['Title','Cast and crew','genres']]
df_2018
```

```python
def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0]
        
df_2018['director_name'] = df_2018['Cast and crew'].map(lambda x: get_director(x))
```

```python
def get_actor1(x):
    return ((x.split("screenplay); ")[-1]).split(", ")[0])
    
df_2018['actor_1_name'] = df_2018['Cast and crew'].map(lambda x: get_actor1(x))
```

```python
def get_actor2(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 2:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[1])
        
df_2018['actor_2_name'] = df_2018['Cast and crew'].map(lambda x: get_actor2(x))
```

```python
def get_actor3(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 3:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[2])
        
df_2018['actor_3_name'] = df_2018['Cast and crew'].map(lambda x: get_actor3(x))
```

```python
df_2018
```

```python
df_2018 = df_2018.rename(columns={'Title':'movie_title'})
df_2018 = df_2018[['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
df_2018
```

```python
df_2018.isnull().sum()
```

```python
df_2018['actor_2_name'] = df_2018['actor_2_name'].replace(np.nan, 'unknown')
df_2018['actor_3_name'] = df_2018['actor_3_name'].replace(np.nan, 'unknown')
df_2018['genres'] = df_2018['genres'].replace(np.nan, 'unknown')
```

```python
df_2018['movie_title'] = df_2018['movie_title'].str.lower()
df_2018['comb'] = df_2018['actor_1_name'] + ' ' + df_2018['actor_2_name'] + ' '+ df_2018['actor_3_name'] + ' '+ df_2018['director_name'] +' ' + df_2018['genres']
df_2018
```
